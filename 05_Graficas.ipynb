```
import matplotlib.pyplot as plt
import numpy as np
t = np.linspace(0,2*np.pi,100)

x = 16*np.sin(t)**3
y = 13*np.cos(t) - 5*np.cos(2*t) - 2*np.cos(3*t) - np.cos(4*t)

# Plot line with RGB tuple (red=1, green=0.2, blue=0.5)
# and 20pt line width
plt.plot(x,y,c=(1,0.2,0.5),lw=20)

# Add features to our figure
plt.text(0,0,'I love Python',horizontalalignment='center',fontsize=28,color='c')
plt.axis('equal')
plt.axis('off')

plt.show()
```

```
# First create some toy data:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

# Create just a figure and only one subplot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')

# Create two subplots and unpack the output array immediately
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)

# Create four polar axes and access them through the returned array
fig, axs = plt.subplots(2, 2)
axs[0, 0].hist(np.random.randn(100),bins=20)
axs[0, 1].plot(np.random.randn(50).cumsum(),'k--')
axs[1, 0].pie(np.random.randn(10))
axs[1, 1].scatter(np.arange(30),np.arange(30)+3*np.random.randn(30))

plt.show()
```

```
# Create some mock data
t = np.arange(0.01, 10.0, 0.01)
data1 = np.exp(t)
data2 = np.sin(2 * np.pi * t)

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('time (s)')
ax1.set_ylabel('exp', color=color)
ax1.plot(t, data1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('sin', color=color)  # we already handled the x-label with ax1
ax2.plot(t, data2, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```

```
x=np.linspace(-10,30,200)
y=np.pi*x**2*(30-x)/3-1000

fig, ax = plt.subplots()

lados=['top','right','bottom','left']
for lado in lados:
  ax.spines[lado].set_position('zero')
  ax.spines[lado].set_color('y')

plt.plot(x,y,lw=2)
plt.fill_between(x,y,where=(x>=10)&(x<=25),color='c' )
plt.text(12,1000,'$\int_a^bf(x)dx$',color='w',fontsize=20)
plt.grid(linestyle='--')
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return np.cos(x)+np.sin(3*x)

x=np.linspace(0,5)
y=f(x)
fig=plt.figure(figsize=(14,6))
ax=plt.gca()
plt.plot(x,y,linewidth=2)
plt.grid(linestyle='--', linewidth=0.9, alpha=0.35)
plt.axvspan(xmin=1, xmax=2, facecolor='g', alpha=0.1)
plt.axhspan(ymin=f(1), ymax=f(2), facecolor='b', alpha=0.1)
plt.axhline(f(3.6),0,5,linestyle='-.',color='r')
plt.axvline(3.6,f(3.6),2,linestyle='--',color='m')
plt.show()
```

```
fig, ax = plt.subplots(figsize=(12,8))
x = np.linspace(-0.50, 1., 1000)

# Outer Plot
ax.plot(x, x**2)
ax.plot(x, np.sin(x))
ax.set(xlim=(-0.5, 1.0), ylim=(-0.5,1.2))
fig.tight_layout()
plt.grid(linestyle='--')

# Inner Plot
inner_ax = fig.add_axes([0.2, 0.55, 0.35, 0.35]) # x, y, width, height
inner_ax.plot(x, x**2,'--')
inner_ax.plot(x, np.sin(x),'-.')
inner_ax.set(title='Acercamiento', xlim=(0.87,0.88),ylim=(0.75,0.78))
inner_ax.grid(linestyle='--')
ax.set_title("Grafica principal", fontsize=20)
plt.show()
```

```
import plotly.graph_objects as go

x=np.linspace(-5,5)
y=x*np.cos(x)

fig=go.Figure()
fig.add_trace(go.Scatter(x=x,y=y,name='Cos'))
fig.update_layout(showlegend=True)
fig.show()
```

```
import plotly.graph_objects as go
import numpy as np

x=np.linspace(-5,5)
y1=x*np.cos(x)
y2=x*np.sin(x)

fig=go.Figure()

fig.add_trace(go.Scatter(x=x,y=y1,name='Cos',mode='markers+lines'))
fig.add_trace(go.Scatter(x=x,y=y2,name='Sin',mode='markers'))

fig.update_layout(showlegend=True,
    title="Mi grafica",
    xaxis_title="Eje X",
    yaxis_title="Eje Y")
fig.show()
```

```
import plotly.graph_objects as go
import numpy as np

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig=go.Figure()

fig.add_trace(go.Surface(x=X,y=Y,z=Z))

fig.update_layout(scene = dict(
                    xaxis_title='Eje X',
                    yaxis_title='Eje Y',
                    zaxis_title='Eje Z'),
                  title='Mi grafica')

fig.show()
```

```
import plotly.graph_objects as go
import numpy as np

x=np.array ([300 ,350 ,400 ,450 ,500]) #temperatura
y=np.array ([150 ,200 ,250 ,300])      #presion
#entalpia
z=np.array ([[3073.3 ,3174.7 ,3277.5 ,3381.7 ,3487.6] ,\
              [3072.1 ,3173.8 ,3276.7 ,3381.1 ,3487.0] ,\
              [3070.9 ,3172.8 ,3275.9 ,3380.4 ,3486.5] ,\
              [3069.7 ,3171.9 ,3275.2 ,3379.8 ,3486.0]])

X, Y = np.meshgrid(x, y)

fig=go.Figure()

fig.add_trace(go.Surface(x=X,y=Y,z=z))
fig.update_layout(scene = dict(
                    xaxis_title='Temperatura',
                    yaxis_title='Presion',
                    zaxis_title='Entalpia'),
                  title='Vapor sobre calentado')
fig.show()
```
