```
# First create some toy data:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

# Create just a figure and only one subplot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')

# Create two subplots and unpack the output array immediately
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)

# Create four polar axes and access them through the returned array
fig, axs = plt.subplots(2, 2)
axs[0, 0].hist(np.random.randn(100),bins=20)
axs[0, 1].plot(np.random.randn(50).cumsum(),'k--')
axs[1, 0].pie(np.random.randn(10))
axs[1, 1].scatter(np.arange(30),np.arange(30)+3*np.random.randn(30))

plt.show()
```

```
# Create some mock data
t = np.arange(0.01, 10.0, 0.01)
data1 = np.exp(t)
data2 = np.sin(2 * np.pi * t)

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('time (s)')
ax1.set_ylabel('exp', color=color)
ax1.plot(t, data1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('sin', color=color)  # we already handled the x-label with ax1
ax2.plot(t, data2, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```

```
x=np.linspace(-10,30,200)
y=np.pi*x**2*(30-x)/3-1000

fig, ax = plt.subplots()

lados=['top','right','bottom','left']
for lado in lados:
  ax.spines[lado].set_position('zero')
  ax.spines[lado].set_color('y')

plt.plot(x,y,lw=2)
plt.fill_between(x,y,where=(x>=10)&(x<=25),color='c' )
plt.text(12,1000,'$\int_a^bf(x)dx$',color='w',fontsize=20)
plt.grid(linestyle='--')
plt.show()
```

```
x=np.linspace(0,5,100)
y1=x**2-5
y2=10*np.sin(x)

fig, ax = plt.subplots()

lados=['top','right','bottom','left']
for lado in lados:
  ax.spines[lado].set_position('zero')
  ax.spines[lado].set_color('y')

plt.plot(x,y1,x,y2)
plt.fill_between(x,y1,y2,where=(x>=3)&(x<=4.5),color='c')
plt.grid(linestyle='--')
plt.show()
```

```
x = np.linspace(0, 10, 11)
y = [3.9, 4.4, 10.8, 10.3, 11.2, 13.1, 14.1,  9.9, 13.9, 15.1, 12.5]

# fit a linear curve an estimate its y-values and their error.
a, b = np.polyfit(x, y, deg=1)
y_est = a * x + b
y_err = x.std() * np.sqrt(1/len(x) +
                          (x - x.mean())**2 / np.sum((x - x.mean())**2))

fig, ax = plt.subplots()
ax.plot(x, y, 'o', color='tab:green')
ax.set_ylim(0,18)
fig.savefig('1.png')
ax.plot(x, y_est, '-.')
ax.set_ylim(0,18)
fig.savefig('2.png')
ax.fill_between(x, y_est - y_err, y_est + y_err, alpha=0.2)
ax.set_ylim(0,18)
fig.savefig('3.png')

plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return np.cos(x)+np.sin(3*x)

x=np.linspace(0,5)
y=f(x)
fig=plt.figure(figsize=(10,6))
ax=plt.gca()
plt.plot(x,y,linewidth=2)
plt.grid(linestyle='--', linewidth=0.9, alpha=0.35)
plt.axvspan(xmin=1, xmax=2, facecolor='r', alpha=0.1)
plt.axhspan(ymin=f(1), ymax=f(2), facecolor='b', alpha=0.1)
plt.axhline(f(3.6),0,5,linestyle='-.',color='r')
plt.axvline(3.6,f(3.6),2,linestyle='--',color='m')
plt.show()
```

```
plt.figure(figsize=(10,7), dpi=80)
x = np.linspace(0, 2*np.pi, 1000)
sine = plt.plot(x,np.sin(x))
cosine = plt.plot(x,np.cos(2*x))
sine_2 = plt.plot(x,np.sin(x+.5))
cosine_2 = plt.plot(x,np.cos(x+.5))
plt.gca().set(ylim=(-1.25, 1.5), xlim=(-.5, 7))
plt.title('Ejemplo de leyendas', fontsize=18)

# Modify legend
plt.legend([sine[0], cosine[0], sine_2[0], cosine_2[0]],   # plot items
           ['sin(x)', 'cos(2x)', 'sin(x+0.5)', 'cos(x+0.5)'],  
           frameon=True,                                   # legend border
           framealpha=1,                                   # transparency of border
           ncol=2,                                         # num columns
           shadow=True,                                    # shadow on
           borderpad=1,                                    # thickness of border
           title='Senos y Cosenos',                        # title
           loc='best')                                     # localizacion
plt.show()
```

```
fig, ax = plt.subplots(figsize=(12,8))
x = np.linspace(-0.50, 1., 1000)

# Outer Plot
ax.plot(x, x**2)
ax.plot(x, np.sin(x))
ax.set(xlim=(-0.5, 1.0), ylim=(-0.5,1.2))
fig.tight_layout()
plt.grid(linestyle='--')

# Inner Plot
inner_ax = fig.add_axes([0.2, 0.5, 0.40, 0.40]) # x, y, width, height
inner_ax.plot(x, x**2)
inner_ax.plot(x, np.sin(x))
inner_ax.set(title='Acercamiento', xlim=(0.84,0.9),ylim=(0.7,0.8))
inner_ax.grid(linestyle='--')
ax.set_title("Grafica principal", fontsize=20)
plt.show()
```

```
# Scatterplot with varying size and color of points
import pandas as pd
midwest = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/midwest_filter.csv")

# Plot
fig = plt.figure(figsize=(14, 7), dpi= 80, facecolor='w', edgecolor='k')    
plt.scatter('area', 'poptotal', data=midwest, s='dot_size', c='popdensity', cmap='Reds', edgecolors='red', linewidths=.5)
plt.title("Bubble Plot of PopTotal vs Area\n(color: 'popdensity' & size: 'dot_size' - both are numeric columns in midwest)", fontsize=16)
plt.xlabel('Area', fontsize=18)
plt.ylabel('Poptotal', fontsize=18)
plt.colorbar()
plt.show()    
```

```
# Generate normal distributed sample points in 2-D
x, y = np.random.multivariate_normal([0, 0], [[1, 1], [1, 2]], 3000).T
# Define subplot styles
fig = plt.figure(figsize=(10, 10))
grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
ax = fig.add_subplot(grid[:-1, 1:])
ax_x = fig.add_subplot(grid[-1, 1:])
ax_y = fig.add_subplot(grid[:-1, 0])
# Plot the data to subplots
ax.scatter(x, y, s=5, alpha=0.5)
ax_x.hist(x, 40, orientation='vertical')
ax_y.hist(y, 40, orientation='horizontal')
ax_x.invert_yaxis()
ax_y.invert_xaxis()
plt.show()
```

```
fig,ax = plt.subplots()
ax.axis('equal')
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
ax.pie(students,
       labels = langs,autopct='%1.2f%%',
       startangle=90,
       shadow=True)
#draw circle
centre_circle = plt.Circle((0,0),0.75,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')  
plt.tight_layout()
plt.show()
```
