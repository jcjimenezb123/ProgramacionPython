```python
import sympy as sp
import numpy as np
```

```python
x,y = sp.symbols('x y')
f = x**2+x-1
sp.plot(f,(x,-2,3))
```

```python
sp.plot_parametric((sp.cos(x),sp.sin(x)),aspect_ratio=(1,1))
```

```python
sp.plot_parametric((x,sp.acos(1-np.abs(x))-np.pi),(x,sp.sqrt(1-(np.abs(x)-1)**2)),axis=False)
```

```python
sp.plot((sp.acos(1-np.abs(x))-np.pi),(sp.sqrt(1-(np.abs(x)-1)**2)),axis=False)
```

```python
import matplotlib.pyplot as plt
```

```python
f = lambda x:np.arccos(1-np.abs(x))-np.pi
g = lambda x:np.sqrt(1-(np.abs(x)-1)**2)
x = np.linspace(-2,2,101)
fig, ax = plt.subplots()
ax.plot(x,f(x),x,g(x))
ax.set_axis_off()
```
