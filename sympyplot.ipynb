```python
import sympy as sp
import numpy as np
```

```python
x,y = sp.symbols('x y')
f = x**2+x-1
sp.plot(f,(x,-2,3))
```

```python
sp.plot_parametric((x,2*x))
```

```python
sp.plot_parametric((x,sp.acos(1-np.abs(x))-np.pi),(x,sp.sqrt(1-(np.abs(x)-1)**2)),axis=False)
```
